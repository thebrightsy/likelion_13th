```python
8/11 KBS의뢰 케이스탯리서치 여론 조사 결과
70%접종이후 방역정책 전환 필요한가? 
import matplotlib.pyplot as plt

ratio = [60.7, 39.3]
labels = ['찬성', '반대']

explode = [0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```
